```python
# Plot histograms for all numerical features
df.hist(figsize=(20, 15), bins=20, edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

```python
# Select numerical features from the DataFrame
numeric_features = df.select_dtypes(include=['number']).columns.tolist()

# Determine number of rows and columns for subplots
num_features = len(numeric_features)
rows = (num_features // 2) + (num_features % 2)  # Two plots per row

plt.figure(figsize=(15, 5 * rows))  # Adjust figure size based on the number of rows
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.02)

# Loop through numerical features and plot
for i, feature in enumerate(numeric_features, 1):
    plt.subplot(rows, 2, i)  # Create subplot grid
    sns.kdeplot(x=df[feature], color='blue', fill=True)  # Kernel Density Estimation plot
    plt.xlabel(feature, fontsize=12)
    plt.title(f'Distribution of {feature}', fontsize=14)
    plt.tight_layout()

plt.show()
```

```python
# Boxplot for numerical features
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.title("Boxplots for Numerical Features")
plt.show()
```

```python
import scipy.stats as stats
# Define target variable (Total Precipitation)
target_var = "tp"  # Adjust if needed

# Ensure full output visibility
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

# Calculate Correlations
correlations = df.corr()[target_var].sort_values(ascending=False)
spearman_corr = df.corr(method="spearman")[target_var].sort_values(ascending=False)
kendall_corr = df.corr(method="kendall")[target_var].sort_values(ascending=False)

# Display Full Correlations
print("\n📊 Pearson Correlation with Precipitation (Linear Relationship)")
print(correlations.to_string(), "\n")

print("\n📊 Spearman Correlation with Precipitation (Monotonic Relationship)")
print(spearman_corr.to_string(), "\n")

print("\n📊 Kendall Correlation with Precipitation (Ranking Relationship)")
print(kendall_corr.to_string(), "\n")

# Visualize Correlations
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

```
